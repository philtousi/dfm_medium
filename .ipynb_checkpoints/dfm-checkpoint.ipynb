```python
import pandas as pd
import numpy as np
import pandas_datareader as web
import datetime as dt
import random
import statsmodels.tsa.api as sm
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11,5)
plt.rcParams['savefig.dpi'] = 300
from sklearn.metrics import mean_squared_error as mse
import warnings
warnings.filterwarnings('ignore')
```

START = dt.datetime(2001, 11, 1)
END = dt.datetime(2022, 11, 30)
random.seed(61)

```python
df = pd.read_csv('./LCDMA_October_2022/balanced_can_md.csv')
df = df.set_index(pd.DatetimeIndex(df['Date'])).iloc[200:,1:30]
df.describe()
```

```python
num_oos=60
testset = df.iloc[-num_oos:]
iters=300

model = sm.DynamicFactorMQ(endog=df)
results = model.fit(maxiter=iters, disp=True)

dfm_pred = results.predict(start=df.shape[0]-num_oos, end=df.shape[0]-1)
```

```python
mse_gdp_in = mse(dfm_pred['GDP_new'], testset['GDP_new'])
mse_gdp_in
```

```python
plt.plot(dfm_pred['GDP_new'],  label='DFM')
plt.plot(testset['GDP_new'], label='Actual')
plt.legend()
plt.show()
```

```python

h=1
dfm_forecast = pd.DataFrame()

model = sm.DynamicFactorMQ(endog=df)
results = model.fit(maxiter=iters, disp=False)

for i in range(-num_oos,0,h):

    results = results.apply(endog=df.iloc[:i]) # might be
    forecast = results.forecast(steps=h)
    dfm_forecast = pd.concat([dfm_forecast, forecast])
    
```

```python
mse_gdp_oos = mse(dfm_forecast['GDP_new'], testset['GDP_new'])
mse_gdp_oos
```

```python
plt.plot(dfm_forecast['GDP_new'],  label='DFM')
plt.plot(testset['GDP_new'], label='Actual')
plt.legend()
plt.show()
```

```python
results = results.apply(df)
forecast = results.forecast(steps=h)
forecast
```

```python
pd.DataFrame(results.factors.smoothed/results.factors.smoothed.std(), index=df.index).plot(color=['c'])
plt.show()
```
